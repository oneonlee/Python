# 1. 간단한 txt 파일

```python
infile = open('Kennedy.txt','r')
text = infile.read()
word_list = text.split()
word_count_dict = {}
for word in word_list:
    if word in word_count_dict:
        word_count_dict[word] += 1
    else:
        word_count_dict[word] = 1

import matplotlib.pyplot as plt
plt.bar(list(word_count_dict.keys()),list(word_count_dict.values()))
plt.grid()
plt.xticks(rotation=45)
plt.yticks(range(5))
plt.title('Frequency of words')
plt.ylabel('Frequency')
plt.show()

# 참고 : https://appia.tistory.com/172
```

## 2-2. 상위 10개만 나타내도록 코드 수정

```python
infile = open('The Memoirs of Sherlock Holmes by Arthur Conan Doyle.txt','r')
text = infile.read()
word_list = text.split()
word_count_dict = {}
for word in word_list:
    if word in word_count_dict:
        word_count_dict[word] += 1
    else:
        word_count_dict[word] = 1

# 상위 10개만 나타내는 코드
# 딕셔너리의 value값에 따라 내림차순으로 정렬
import operator
sorted_word_list = sorted(word_count_dict.items(), key=operator.itemgetter(1), reverse = True) 

# 내림차순으로 정렬한 list를 바탕으로 상위 10개만 뽑아 dictionary로 만듬
top10_word_dict = {}
i = 0
for tup in sorted_word_list:
    if i < 10:
        top10_word_dict[tup[0]] = tup[1]
        i = i+1
    else:
        break

import matplotlib.pyplot as plt
x = list(top10_word_dict.keys())
y = list(top10_word_dict.values())
plt.bar(x, y)
plt.grid()
plt.xticks(rotation=45)
# plt.yticks(range(5))

# 그래프에 값 표시하는 코드
for i, v in enumerate(x):
    plt.text(v, y[i], y[i],
             fontsize = 9,
             color = "black",
             horizontalalignment = 'center',
             verticalalignment = 'bottom')

plt.title('Frequency of words')
plt.ylabel('Frequency')
plt.show()
```
